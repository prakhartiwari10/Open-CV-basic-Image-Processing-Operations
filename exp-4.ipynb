```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("/home/prakhar/Downloads/West_Coast_New_Zealand.jpg")
```

```python
plt.imshow(img)
```

```python
bit_planes = []
```

```python
for k in range(8):
    bitPlane = []
```

```python
for i in range(8):
    
    mask = np.uint8(1 << i)
    
    bit_plane = np.zeros_like(img)
   
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
           
            pixel = img[y, x]
            
            bit = (pixel & mask) >> i
            
            bit_plane[y, x] = bit * 255  
    bit_planes.append(bit_plane)
```

```python
plt.figure(figsize=(10, 5))
for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title('Bit Plane {}'.format(i))
    plt.axis('off')

plt.tight_layout()
plt.show()
```

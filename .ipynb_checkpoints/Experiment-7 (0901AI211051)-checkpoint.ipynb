```python
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt

img = cv2.imread('C:/Users/lodhi/Documents/Apollo_11_Launch.jpg', cv2.IMREAD_GRAYSCALE)
c = np.array(img.data)

plt.subplot(2, 5, 1)
plt.title('ORIGINAL')
plt.imshow(c, cmap='gray')

finalImage = np.zeros_like(c)

for k in range(8):
    bitPlane = []
    for i in range(c.shape[0]):
        a = []
        for j in range(c.shape[1]):
            if c[i][j] % 2 == 1:
                a.append(255)
            else:
                a.append(0)
            c[i][j] = c[i][j] // 2
        bitPlane.append(a)
    img = np.array(bitPlane)
    for i in range(c.shape[0]):
        for j in range(c.shape[1]):
            if img[i][j] == 255:
                finalImage[i][j] += np.power(2, k)

    plt.subplot(2, 5, k + 2)
    plt.title('BP-' + str(k))
    plt.imshow(img, cmap='gray')

plt.subplot(2, 5, 10)
plt.imshow(finalImage, cmap='gray')
plt.title('RECONSTRUCTED')
plt.show()
```

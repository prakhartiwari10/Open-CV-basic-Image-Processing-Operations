```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Replication Zoom
img = cv2.imread('C:/Users/lodhi/Documents/Apollo_11_Launch.jpg', cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Unable to load image.")
else:
    plt.subplot(1, 2, 1)
    plt.title('ORIGINAL')
    plt.imshow(img, cmap='gray')
    print(img)

    # ROW WISE ZOOMING
    for i in range(0, img.shape[0]*2, 2):
        x = img[:, i:i+1]
        y = x.flatten()
        img = np.insert(img, i, y, 1)

    # COLUMN WISE ZOOMING
    for i in range(0, img.shape[1]*2, 2):  # Corrected the loop range
        x = img[:, i//2:i//2+1].flatten()  # Corrected the indexing and flattening
        img = np.insert(img, i, x, 1)  # Inserting flattened column vector

    print(img)
    plt.subplot(1, 2, 2)
    plt.title('REPLICATION ZOOM')
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
# Read the image
img = cv2.imread('C:/Users/lodhi/Documents/Apollo_11_Launch.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if img is None:
    print("Error: Unable to load image.")
else:
    img = np.array(img)

    # Display the original image
    plt.subplot(1, 2, 1)
    plt.title('ORIGINAL')
    plt.imshow(img, cmap='gray')

    # Perform ROW WISE ZOOMING
    for i in range(1, 2*img.shape[0]-1, 2):
        x = img[:, i-1:i].flatten()
        y = img[:, i:i+1].flatten()
        z = np.add(x, y, dtype=np.int64)
        img = np.insert(img, i, z/2, 1)

    # Perform COLUMN WISE ZOOMING
    for i in range(1, 2*img.shape[0]-1, 2):
        x = img[i - 1:i, :].flatten()
        y = img[i:i + 1, :].flatten()
        z = np.add(x, y, dtype=np.int64)
        img = np.insert(img, i, z/2, 0)

    # Display the interpolated image
    plt.subplot(1, 2, 2)
    plt.title('INTERPOLATION ZOOM')
    plt.imshow(img, cmap='gray')
    plt.show()
```

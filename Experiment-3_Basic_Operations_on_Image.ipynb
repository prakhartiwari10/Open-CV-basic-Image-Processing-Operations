```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread("/home/Downloads/West_Coast_New_Zealand.jpg")
```

```python
plt.imshow(img)
```

```python
img2  = np.ones(img.shape,dtype=np.uint8)*50
```

```python
img_bright = cv2.add(img, img2)
```

```python
plt.imshow(img_bright)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img)
axs[0].axis('off')  # Hide axis

# Plot the second image in the second subplot (right)
axs[1].imshow(img_bright)
axs[1].axis('off')  # Hide axis

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
img_dark = cv2.subtract(img, img2)
```

```python
fig, axs = plt.subplots(1, 3)
axs[0].imshow(img)
axs[0].axis('off')  # Hide axis

# Plot the second image in the second subplot (right)
axs[1].imshow(img_bright)
axs[1].axis('off')  # Hide axis

axs[2].imshow(img_dark)
axs[2].axis('off') 

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
img3 =  np.ones(img.shape)*1.2
img4 = np.ones(img.shape)*0.8
```

```python
image_float = img.astype('float32')
```

```python
high_contrast = (image_float * img3).astype('uint8')
low_contrast = (image_float * img4).astype('uint8')
```

```python
fig, axs = plt.subplots(1, 3)
axs[0].imshow(img)
axs[0].axis('off')  # Hide axis

# Plot the second image in the second subplot (right)
axs[1].imshow(high_contrast)
axs[1].axis('off')  # Hide axis

axs[2].imshow(low_contrast)
axs[2].axis('off') 

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the subplots
plt.show()
```
